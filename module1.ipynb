# combine cusotmer info with customer transaction 

```python
cus_transaction1=cus_transaction.copy()
```

```python
# drop null value before join
cus_transaction1.dropna(how='all')
```

```python
# fill blanks and n/a with NA,
cus_transaction1
cols=["online_order","brand ",'product_line ','product_class ','product_size','standard_cost','product_first_sold_date']
def replace_columns(row):
    for col in cols: 
        if row[col]=="n/a ":
            row[col]=="NA"
        return row 
cus_transaction1.apply(replace_columns,axis=1)
```

```python
# fill non value data with NA 
cus_transaction1["online_order"].fillna("NA",inplace=True)
cus_transaction1["brand"].fillna("NA",inplace=True)
cus_transaction1["product_line"].fillna("NA",inplace=True)
cus_transaction1["product_class"].fillna("NA",inplace=True)
cus_transaction1["product_size"].fillna("NA",inplace=True)
cus_transaction1["standard_cost"].fillna("NA",inplace=True)
cus_transaction1["product_first_sold_date"].fillna("NA",inplace=True)
```
